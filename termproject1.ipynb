```python
import pandas as pd
cnts = pd.read_csv('read-counts.txt', sep='\t', comment='#', index_col=0)
cnts.head(10)
```

```python
cnts['binfo1-datapack1/CLIP-35L33G.bam'].hist(bins=100, range=(0,100))
```

```python
cnts['binfo1-datapack1/RNA-control.bam'].hist(bins=100, range=(0,100))
```

```python
cnts['clip_enrichment'] = cnts['binfo1-datapack1/CLIP-35L33G.bam'] / cnts['binfo1-datapack1/RNA-control.bam']
cnts['rden_change'] = (cnts['binfo1-datapack1/RPF-siLin28a.bam'] / cnts['binfo1-datapack1/RNA-siLin28a.bam']) / (cnts['binfo1-datapack1/RPF-siLuc.bam'] / cnts['binfo1-datapack1/RNA-siLuc.bam'])
cnts.head()
```

```python
from matplotlib import pyplot as plt
import numpy as np

fig, ax = plt.subplots(1, 1, figsize=(5, 5))
ax.scatter(np.log2(cnts['clip_enrichment']),
           np.log2(cnts['rden_change']), s=1)

plt.xlim([-6, 4])
plt.ylim([-2, 2])
```

```python
## RPM 구하기
total_reads_CLIP = cnts['binfo1-datapack1/CLIP-35L33G.bam'].sum()
total_reads_RNA = cnts['binfo1-datapack1/RNA-control.bam'].sum()
total_reads_RPF_Lin = cnts['binfo1-datapack1/RPF-siLin28a.bam'].sum()
total_reads_RNA_Lin = cnts['binfo1-datapack1/RNA-siLin28a.bam'].sum()
total_reads_RPF_Luc = cnts['binfo1-datapack1/RPF-siLuc.bam'].sum()
total_reads_RNA_Luc = cnts['binfo1-datapack1/RNA-siLuc.bam'].sum()
cnts['CLIP-RPM'] = (cnts['binfo1-datapack1/CLIP-35L33G.bam']/ total_reads_CLIP) * 1e6
cnts['RNA-RPM'] = (cnts['binfo1-datapack1/RNA-control.bam']/ total_reads_RNA) * 1e6
cnts['RPF-Lin-RPM'] = (cnts['binfo1-datapack1/RPF-siLin28a.bam']/ total_reads_RPF_Lin) * 1e6
cnts['RNA-Lin-RPM'] = (cnts['binfo1-datapack1/RNA-siLin28a.bam']/total_reads_RNA_Lin) * 1e6
cnts['RPF-Luc-RPM'] = (cnts['binfo1-datapack1/RPF-siLuc.bam']/total_reads_RPF_Luc) * 1e6
cnts['RNA-Luc-RPM'] = (cnts['binfo1-datapack1/RNA-siLuc.bam']/total_reads_RNA_Luc) * 1e6

cnts['clip_enrichment'] = cnts['CLIP-RPM'] / cnts['RNA-RPM']
cnts['rden_change'] = (cnts['RPF-Lin-RPM'] / cnts['RNA-Lin-RPM']) / (cnts['RPF-Luc-RPM'] / cnts['RNA-Luc-RPM'])
cnts.head(20)
```

```python
# NaN값이 존재한다 - 결측치 0으로 바꿈
cnts = cnts.fillna(0)

# inf값이 존재한다 -  NaN값으로 바꾼 후 제거
cnts= cnts.replace([np.inf, -np.inf], np.nan)
cnts.dropna(inplace=True)


from matplotlib import pyplot as plt
import numpy as np

fig, ax = plt.subplots(1, 1, figsize=(5, 5))
ax.scatter(np.log2(cnts['clip_enrichment']),
               np.log2(cnts['rden_change']), s = 1) # dot 크기 조절

#논문처럼 x축, y축 범위 정함
plt.xlim([-6, 4])
plt.ylim([-2, 2])

# 논문처럼 비례관계를 보여주지만 이상한 점들이 많다. 그래서 dataframe을 조절 필요
```

```python
#low read count(<30)와 low ribosome foot prints(<80) 삭제
idx1 = cnts[cnts['binfo1-datapack1/CLIP-35L33G.bam'] < 30].index
cnts.drop(idx1, inplace=True)
idx2 = cnts[cnts['binfo1-datapack1/RPF-siLin28a.bam'] < 80].index
cnts.drop(idx2, inplace=True)
cnts.head(20)

# 다시 그리기

# NaN값이 존재한다 - 결측치 0으로 바꿈
cnts = cnts.fillna(0)

# inf값이 존재한다 -  NaN값으로 바꾼 후 제거
cnts= cnts.replace([np.inf, -np.inf], np.nan)
cnts.dropna(inplace=True)

from matplotlib import pyplot as plt
import numpy as np

fig, ax = plt.subplots(1, 1, figsize=(5, 5))
ax.scatter(np.log2(cnts['clip_enrichment']),
               np.log2(cnts['rden_change']), s = 1) # dot 크기 조절

#논문처럼 x축, y축 범위 정함
plt.xlim([-6, 4])
plt.ylim([-2, 2])
```

```python
import ssl
ssl._create_default_https_context = ssl._create_unverified_context
mouselocal = pd.read_csv('https://hyeshik.qbio.io/binfo/mouselocalization-20210507.txt', sep='\t')
mouselocal.tail(10)
```

```python
types =['None']*len(cnts)

for i in range(len(cnts)):
  for j in range(len(mouselocal)):
    if mouselocal['gene_id'][j] in cnts.index[i]:
      types[i] = mouselocal['type'][j]
      break

cnts['type'] = types
```

```python
nucleus = cnts[cnts['type'] =='nucleus']
integral_membrane = cnts[cnts['type'] =='integral membrane']
cytoplasm = cnts[cnts['type']=='cytoplasm']

plt.scatter(np.log2(nucleus['clip_enrichment']), np.log2(nucleus['rden_change']), color='blue', s=1)
plt.scatter(np.log2(integral_membrane['clip_enrichment']), np.log2(integral_membrane['rden_change']), color='red', s=1)
plt.scatter(np.log2(cytoplasm['clip_enrichment']), np.log2(cytoplasm['rden_change']), color='green', s=1)

plt.xlim([-6, 4])
plt.ylim([-2, 2])
```
